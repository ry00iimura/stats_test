```
# Mann–Whitney test
import numpy as np
import pandas as pd
n = 50
val = np.random.randn(n)
cls = np.random.choice(['A', 'B'], n)
a = pd.DataFrame(dict(cls=cls, val=val))
a.head()
```

```
import pandas as pd
from scipy import stats
stats.mannwhitneyu(a['val'][a['cls'] == 'A'], a['val'][a['cls'] == 'B']) #one-side
stats.mannwhitneyu(a['val'][a['cls'] == 'A'], a['val'][a['cls'] == 'B'], alternative='two-sided') #two-sides
```
